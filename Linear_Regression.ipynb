```python
#The following lines are necessary to use existing libraries
import numpy as np
import matplotlib.pyplot as plt
import sklearn   # This imports the scikit-learn library
```

```python
# The following lines make the training data (x: input, y: output)
# Data is a noise added version of y = 2*x + 1

np.random.seed(0)
n=20   # Number of data points
x=np.linspace(0, 10, n)
y=x*2 + 1 + 1.0*np.random.randn(n)  # *5.0  # Standard deviation 1
print(x)
print(y)

# data visualization
n=20   # Number of data points
x_=np.linspace(0, 10, n)
y_=x_*2 + 1
plt.plot(x_,y_)
plt.plot(x,y, 'o')
```

```python
# The following lines run linear regression by using existing codes from library
from sklearn.linear_model import LinearRegression
model=LinearRegression( )
model.fit(x[:,np.newaxis], y)
print(model.coef_)
print(model.intercept_)

# data visualization
y_reg = model.coef_ * x_ + model.intercept_;

plt.plot(x_,y_)  # original data without noise
plt.plot(x_, y_reg) # line estimated from the regression (with noise)
plt.plot(x,y, 'o') # data points used

```
